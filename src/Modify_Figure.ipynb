```python
import matplotlib.pyplot as plt
import pickle
import os
import numpy as np
import pandas as pd
```

## Sector Analysis

```python
n=1
sector_figure_folder = "../data/figure_data/sector_analysis"
save = False

for fig_name in os.listdir(sector_figure_folder):
    fig_path = "%s/%s" % (sector_figure_folder,fig_name)
    fig_saved = pickle.load(open(fig_path, 'rb'))
    
    df_NN = fig_saved.axes[0].lines[0].get_data()[1]
    df_linear = fig_saved.axes[0].lines[1].get_data()[1]
    df_PLS_NN_50 = fig_saved.axes[0].lines[2].get_data()[1]
    
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.set_title(fig_saved.axes[0].get_title())
    
    OLS_label='OLS: ' + str(round(np.mean(df_linear),3)) +  ' (' + str(round(np.std(df_linear),3)) + ')'
    NN_label='NN: ' + str(round(np.mean(df_NN),3)) +  ' (' + str(round(np.std(df_NN),3)) +  ')'
    PLS_NN_label='PLS_NN: ' + str(round(np.mean(df_PLS_NN_50),3)) +  ' (' + str(round(np.std(df_PLS_NN_50),3)) +  ')'
    
    ax.plot(range(0,len(df_NN)), df_NN, 'o', ls='--', ms=2, color='red',linewidth=0.75, label=NN_label)
    ax.plot(range(0,len(df_linear)), df_linear, 'o', ls='--', ms=2, color='blue',linewidth=0.75, label=OLS_label)
    ax.plot(range(0,len(df_PLS_NN_50)), df_PLS_NN_50, 'o', ls='--', ms=2, color='green',linewidth=0.75, label=PLS_NN_label)
    
    ax.set_ylim(fig_saved.axes[0].get_ylim())
    ax.set_xticks(fig_saved.axes[0].get_xticks())
    ax.set_xticklabels(fig_saved.axes[0].get_xticklabels())
    ax.tick_params(axis='x', rotation=50)
    ax.axvline(x=90, linestyle='--', color='black')
    ax.legend(loc='upper left')#, bbox_to_anchor=(1, 0.5))
    
    if save:
        name_png = "../figure/factor_tilts_hi/%s_%s.png" % (fig_name[:10], "modified")
        fig.savefig(name_png, format='png',dpi=1200,bbox_inches = "tight")
    
        name_pickle = "../data/figure_data/factor_analysis/%s_%s.fig.pickle" % (fig_name[:10], "modified")
        pickle.dump(fig, open(name_pickle, 'wb')) # This is for Python 3 - py2 may need `file` instead of `open`
```
